```python

vals = []
names = []
first = True
for aa in tf.train.summary_iterator(efile):
    if not len(aa.summary.value):
        continue
    if first:
        for ndx,bb in enumerate(aa.summary.value):
            vals.append([])
            names.append(bb.tag)
        first = False
    for ndx,bb in enumerate(aa.summary.value):
        vals[ndx].append(bb.simple_value)
```
